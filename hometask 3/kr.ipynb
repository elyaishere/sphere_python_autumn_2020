У какого города суммарно наибольшее число успешных заказов (посмотрела и ручками ввела в форму 24).

```python
df.city_id.unique()
print(df[df.city_id == 23].successful_orders.sum())
print(df[df.city_id == 25].successful_orders.sum())
print(df[df.city_id == 26].successful_orders.sum())
print(df[df.city_id == 24].successful_orders.sum())
```

Рестораны какой специализации имеют самую высокую долю неуспешных заказов (все заказы по конкретному ресторану складываются из успешных, неуспешных заказов и перезаказов)?

```python
df['total_orders'] = df['fail_orders'] + df['successful_orders']
k  = df.groupby('spec').fail_orders.sum()/df.groupby('spec').total_orders.sum()
k.sort_values(ascending = False)[[0]].index[0]
```

У какого ресторана был наибольший абсолютный рост числа успешных заказов в августе по сравнению с июлем?

```python
k1 = pd.DataFrame(df[(df.date > '2019-07-00') & (df.date < '2019-08-00')].groupby('chain_id').successful_orders.sum())
k2 = pd.DataFrame(df[(df.date > '2019-08-00') & (df.date < '2019-09-00')].groupby('chain_id').successful_orders.sum())
k3 = k1.merge(k2, how = 'inner', on = 'chain_id')
k3['diff'] = k3['successful_orders_y'] - k3['successful_orders_x']
k3.sort_values(['diff'])
```
